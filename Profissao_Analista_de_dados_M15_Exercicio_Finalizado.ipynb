## 1\. Preços `outliers` de diamante




```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
with sns.axes_style('whitegrid'):
    grafico = sns.boxplot(x=data["price"])
    grafico.set(title="Preços outliers", xlabel = 'Preço (USD)')
```

**Insight do gráfico 1**: ...


*   Os preços mais comuns estão entre 800 e 5200 aproximadamente.
*   A partir de aproximadamente 12000 os preços começam a ficar mais distantes um dos outros.
*   Os diamantes mais baratos são mais comuns, enquanto os mais caros são mais raros.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head(10000)
```

```
# gráfico do exercício 2
diamonds = data[["carat", "price", "clarity"]]
diamonds.head()

with sns.axes_style('whitegrid'):
    grafico = sns.scatterplot(data= diamonds, x="carat", y="price", hue="clarity", palette='bright' )
    grafico.set(title = "Correlação entre o preço do diamante com seu peso, agrupados por transparência", xlabel = "Peso", ylabel = "Preço");
    grafico.get_legend().set_title("Transparência")
```

**Insight do gráfico 2**: ...

*   A proporção de aumento do preço em relação peso, aumenta de acordo com a transparência.
*  Praticamente todas as transparencias apresentam crescimento parecido, com exeção da I1, que está mais distante da curva de crescimento das outras transparencias.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()

```

```
# gráfico do exercício 3

cars = data[['weight']]
cars

with sns.axes_style('whitegrid'):

    grafico = sns.histplot(data=cars, x='weight', kde=True)
    grafico.set(title='Distribuição contínua aproximada do peso (weight) de carros',xlabel= "Peso", ylabel = 'Contagem');

```

**Insight do gráfico 3**: ...


*   O peso mais comum é de aproximadamente entre 2000 e 2400
*   O peso mais incomum é de aproximadamente 5000
*   Há um crescimento no preso que começa em 1500 até aproximadamente 2000, mas depois há uma queda constante.
