## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1
diamonds = data[["cut", "price"]].groupby("cut").agg("mean").reset_index()
diamonds
```

```
grafico = sns.barplot(data= diamonds, x='cut', y='price', ci = None, palette='dark')
grafico.set(title = 'Preço do diamente por tipo de corte', xlabel = 'Corte', ylabel = 'Preço médio(USD)');
```

**Insight do gráfico 1**: ...


*   Os cortes Good e Very Good tem preços semelhantes.
*   Ideal é o mais barato entre os cortes

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```

flights = data.loc[data['month']=='Dec', ['year', 'passengers']].reset_index()
flights
```

```
# gráfico do exercício 2


with sns.axes_style('whitegrid'):

    grafico = sns.lineplot(data=flights,x='year', y='passengers',palette='dark')
    grafico.set(title='Passageiros em Dezembro por Ano', xlabel = 'Ano', ylabel = 'Numero de passageiros');
```

**Insight do gráfico 2**: ...

*   Houve um crescimento no número de passageiros em dezembro a cada ano que passou.
*   Entre 1957 e 1958 houve pouco crescimento no número de passageiros

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")

with sns.axes_style('whitegrid'):
    grafico = sns.lineplot(data=flights, x='month', y='passengers', hue='year', palette='dark')
    grafico.set(title = 'Passageiros por mês entre 1949 e 1959', xlabel = 'Mês', ylabel = 'Passageiros');
    grafico.get_legend().set_title("Ano")
```

**Insight do gráfico 3**: ...

*   Os meses de julho e agosto possuem o maior numero de passageiros, em todos os anos.
*   Em todos os anos há uma queda de passageiros no mês de Fevereiro e Novembro
*   Cada ano tem um aumento significativo de passageiros em relação ao ano anterior
